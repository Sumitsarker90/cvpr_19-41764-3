## Tanh Function

```python
# Tanh(x) = (e^x-e^-x)
def  tanh(x):
    z= (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return z

output=tanh(xSeries)
plt.plot(xSeries, output)
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

# Elu Activation function


```python
import numpy as np

def elu(x, alpha=1.0):
    # Compute the ELU activation function
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
output3= elu(xSeries)

plt.plot(xSeries, output3)
plt.title('ELU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

## Selu Activation Function

```python
import numpy as np

def selu(x, alpha=1.67326, scale=1.0507):
    
    pos_scale = scale*np.maximum(0, x)
    neg_scale = scale*(alpha*np.exp(x)-alpha)
    return np.where(x > 0, pos_scale, neg_scale)

import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
output4= selu(xSeries)

plt.plot(xSeries, output4)
plt.title('SELU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

## Step Function
